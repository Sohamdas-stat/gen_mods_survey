# Generate data

```python
from sklearn.datasets import make_moons

X, z = make_moons(n_samples=100000, noise=0.1, random_state=42) # can adjust the noise level

# save the data
import numpy as np
np.save("./data/X.npy", X)
np.save("./data/z.npy", z)

# Load data
X = np.load('./data/X.npy')
z = np.load('./data/z.npy')
```

# Make scatter plot

```python
import matplotlib.pyplot as plt

plt.scatter(X[z == 0, 0], X[z == 0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X[z == 1, 0], X[z == 1, 1], color='blue', marker='o', alpha=0.5)
plt.show()
```
